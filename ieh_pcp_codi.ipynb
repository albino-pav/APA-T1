```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import sounddevice as sd
```

```python
T = 2.5
fm = 8000
fx = 1000
A = 4
pi = np.pi
L = int(fm * T)
Tm = 1/fm
t=Tm*np.arange(L) 
x = A * np.cos(2 * pi * fx * t)
sf.write('so_exemple1.wav', x, fm)
```

```python
Tx = 1/fx
Ls = int(fm * 5 * Tx)
plt.plot(t[: Ls], x[: Ls])
```

```python
from numpy.fft import fft
N = 5000
X = fft(x[: Ls], N)
k = np.arange(N)
fig, axs = plt.subplots(2, 1)
axs[0].plot(k, abs(X))
axs[1].plot(k, np.unwrap(np.angle(X)))

fig.suptitle("transformada Fourier del senyal")
axs[0].set_title("Modul")
axs[1].set_title("Fase")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import sounddevice as sd
```

```python
T = 2.5
fm = 8000
fx1 = 4000
fx2 = 200
A = 4
pi = np.pi
L = int(fm * T)
Tm = 1/fm
t=Tm*np.arange(L) 
x1 = A * np.cos(2 * pi * fx1 * t)
x2 = A * np.cos(2 * pi * fx2 * t)
sf.write('so_1.wav', x1, fm)
sf.write('so_2.wav', x2, fm)
```

```python
Tx1 = 1/fx1
Tx2 = 1/fx2
Ls1 = int(fm * 5 * Tx1)
Ls2 = int(fm * 5 * Tx2)

plt.figure(0)                             
plt.plot(t[0:Ls1], x1[0:Ls1])             
plt.show()                                

plt.figure(1)                             
plt.plot(t[0:Ls2], x2[0:Ls2])              
plt.show() 
```

```python
plt.plot(x1[0:Ls1])                      
plt.show()                               
                            
plt.plot(x2[0:Ls2])                      
plt.show()                               
```

```python
N = 5000                                 
X1 = np.fft.fft(x1, N)                   
X2 = np.fft.fft(x2, N)                  

plt.figure(4)                           
plt.subplot(2,1,1)                       
plt.plot(np.arange(N), np.abs(X1))      
                       
plt.subplot(2,1,2)                      
plt.plot(np.arange(N), abs(X2))         
                  
```
